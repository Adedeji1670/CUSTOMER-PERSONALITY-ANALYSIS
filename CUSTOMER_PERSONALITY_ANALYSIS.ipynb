```python
#load the libaries
import pandas as pd
import numpy as np
# data visulization
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dt_df = pd.read_csv('marketing_campaign.csv', sep='\t')
```

```python
dt_df.head(4)
```

```python
dt_df.shape
```

```python
dt_df.info()
```

```python
dt_df.isnull().sum()
```

```python
dt_df.fillna({'Income': dt_df['Income'].mean()}, inplace=True)
```

```python
dt_df.isnull().sum()
```

```python
dt_df.describe()
```

```python
dt_df.columns
```

```python
# get customers age from the year_birth
dt_df['age'] = 2024- dt_df['Year_Birth']
```

```python
# converting Dt_customer to date datatype
dt_df = dt_df.astype(
    {'Dt_Customer': 'datetime64[ns]'}
)
```

```python
dt_df.head(3)
```

```python
dt_df.info()
```

```python
# get year as customers column
dt_df['year_as_customer'] = [2024 - x.year for x in dt_df['Dt_Customer']]
```

```python
dt_df.head(2)
```

```python
# dropping redundants columns
dt_df.drop(columns = ['ID','Year_Birth','Dt_Customer'], inplace= True)
```

```python
dt_df.head(2)
```

```python
dt_df.Education.value_counts()
```

#### Here come an Exploratory Data Analysis

```python
fig, axes = plt.subplots(nrows=1,ncols=2, figsize=(12,4))
sns.barplot(data=dt_df, x = 'Education', y = 'Income', ax = axes[0], color='green')
sns.barplot(data=dt_df, x = 'Marital_Status', y = 'Income', ax = axes[1])
axes[0].set_title('Income By Eduaction')
axes[1].set_title('Income By Marital_Status')
plt.suptitle('INCOME BEWTEEN EDUCATION AND MARITAL STATUS')
```

```python
fig, axes = plt.subplots(2,2, figsize = (20,10))
sns.regplot(data = dt_df, x = 'age', y = 'Income', ax=axes[0,0])
sns.regplot(data = dt_df, x = 'year_as_customer', y = 'Income', ax=axes[0,1])
sns.regplot(data = dt_df, x = 'MntWines', y = 'Income', ax=axes[1,0])
sns.regplot(data = dt_df, x = 'MntFruits', y = 'Income', ax=axes[1,1])
axes[0,0].set_title('Income vs Age')
axes[0,1].set_title('Income by Year as customer')
axes[1,0].set_title('Income vs Age')
axes[1,1].set_title('Income by Year as customer')

fig.tight_layout(pad=3)
```

```python
# showing correlation between features to select features for ML model
corr = dt_df.select_dtypes(exclude='object').corr()
corr
```

```python
# correlation heatmap
plt.figure(figsize=(30,30))
sns.heatmap(corr, annot=True,cmap='coolwarm')
```
